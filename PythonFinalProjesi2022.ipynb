```python
# 2003-2021 yılları arasında aylara göre tüketici fiyat endeks rakamlarını gösterir(Kaynak:TÜİK , 'https://data.tuik.gov.tr/Bulten/Index?p=Tuketici-Fiyat-Endeksi-Aralik-2021-45789')
file ='Tüketici Fiyat Endeksi.xls'
df = pd.read_excel(file)
df
```

```python
from bokeh.io import output_file, show
from bokeh.layouts import row
from bokeh.plotting import figure

output_file("layout.html")
c1 = figure(width=250, plot_height=250, title='Ocak')
c1.circle(df['Unnamed: 0'], df['Ocak'], size=10, color='navy', alpha=0.5)
c2 = figure(width=250, height=250,title='Şubat',x_range=c1.x_range,y_range=c1.y_range) #x ve y aksisler paylaşılıyor
c2.triangle(df['Unnamed: 0'], df['Şubat'],size=10, color='red', alpha=0.5)
c3 = figure(width=250, height=250,title='Mart',x_range=c1.x_range,y_range=c1.y_range) #sadece x aksis paylaşılıyor
c3.square(df['Unnamed: 0'], df['Mart'],size=5, color='green', alpha=0.5)
#Aksislerin paylaşılma nedeni, üç grafikten birinin üzerinde birşeyler yaptığımızda diğer grafiklerde de aynı tepkiyi verecek.
```

```python
pencere=Tk()
pencere.geometry("1700x900") 
#pencere.configure(background='#81e6d9')
canvas = Canvas(pencere, width = 1700, height = 900)  
canvas.pack()  
         
python_img=PhotoImage(file="python.png")
canvas.create_image(250,0,anchor=NW,image=python_img)


barplot2018=PhotoImage(file="2018yilibarplot.png")
barplot2019=PhotoImage(file='2019yilibarplot.png')
barplot2020=PhotoImage(file='2020yilibarplot.png')
barplot2021=PhotoImage(file="2021yilibarplot.png")

bar_ocak=PhotoImage(file="bar_ocak.png")
bar_haziran=PhotoImage(file="bar_haziran.png")
bar_aralik=PhotoImage(file="bar_aralik.png")



def mesaj():
    messagebox.showinfo(title="Başarılı", message="Seçim başarılı")
    
def mesaj2():
    messagebox.showinfo(title="Not", message="Dosya masaüstünüze layout.html olarak kaydedilmiştir.")

def mesaj3():
    messagebox.showinfo(title="Not", message="Dosya masaüstünüze layout2.html olarak kaydedilmiştir.")      
    
def mesaj4():
    messagebox.showinfo(title="Not", message="Dosya masaüstünüze layout3.html olarak kaydedilmiştir.")   
    
    
def mesaj5():
    messagebox.showinfo(title="Not", message="Dosya masaüstünüze layout4.html olarak kaydedilmiştir.") 




def text_bilgisi_al():
    global deger
    if veri_text.get("1.0","end-1c")=="Tüketici Fiyat Endeksi.xls":
        deger=+1
        mesaj()


def interaktif1():
    global count
    if count==0: 
        if deger==1:
            show(row(c1, c2, c3))
            count=+1

        
    elif count>=1:
        mesaj2()



        
def interaktif2():
    global count2
    if count2==0:
        if deger==1:
            show(row(c4,c5,c6))
            count2=+1

    elif count2>=1:
        mesaj3()
        
def interaktif3():
    global count3
    if count3==0:
        if deger==1:
            show(row(c7,c8,c9))
            count3=+1

    elif count3>=1:
        mesaj4()
        
def interaktif4():
    global count4
    if count4==0:
        if deger==1:
            show(row(c10,c11,c12))
            count4=+1

    elif count4>=1:
        mesaj5()
       
bar_kutu=0

bar_cesitleri=['2021','2020','2019','2018']
aylar=['Ocak','Haziran','Aralık']       

def bar_cizdir():
    bar_deger=bar_kutu.get()
    if bar_deger=='2021':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=barplot2021)
        canvas.create_image(250,0,anchor=NW,image=python_img)
    elif bar_deger=='2020':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=barplot2020)
        canvas.create_image(250,0,anchor=NW,image=python_img)
    elif bar_deger=='2019':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=barplot2019)
        canvas.create_image(250,0,anchor=NW,image=python_img)
    elif bar_deger=='2018':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=barplot2018)
        canvas.create_image(250,0,anchor=NW,image=python_img)
  
def bar_cizdir2():
    bar_deger=bar_kutu2.get()
    if bar_deger=='Ocak':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=bar_ocak)
        canvas.create_image(250,0,anchor=NW,image=python_img)
    elif bar_deger=='Haziran':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=bar_haziran)
        canvas.create_image(250,0,anchor=NW,image=python_img)
    elif bar_deger=='Aralık':
        canvas.delete("all")
        canvas.create_image(200, 250, anchor=NW, image=bar_aralik)
        canvas.create_image(250,0,anchor=NW,image=python_img)



def grafik_düzenle():
    grafik_deger=grafik_kutu.get()
    if deger==1: 
        if grafik_deger=="Yıl içinde mevsimlere göre değişim(bar)":
            mesaj()
            pencere2=Tk()
            global bar_kutu
            bar_label=Label(pencere2,text="Yıl Seç",font="helvetica 12",borderwidth=6)
            bar_label.place(x=0,y=10)

            bar_kutu=Combobox(pencere2,values=bar_cesitleri)
            bar_kutu.place(x=0,y=50)
                
            bar_buton=Button(pencere2,text="Onayla",command=bar_cizdir,font="helvetica 12",borderwidth=6)
            bar_buton.place(x=0,y=150)
        elif grafik_deger=="Ayların yıllara göre değişimi(bar)":
            mesaj()
            pencere3=Tk()
            global bar_kutu2
            bar_label2=Label(pencere3,text="Ay Seç",font="helvetica 12",borderwidth=6)
            bar_label2.place(x=0,y=10)
            
            bar_kutu2=Combobox(pencere3,values=aylar)
            bar_kutu2.place(x=0,y=50)
            
            bar_buton2=Button(pencere3,text="Onayla",command=bar_cizdir2,font="helvetica 12",borderwidth=6)
            bar_buton2.place(x=0,y=150)

            
            
grafik_label=Label(text="Grafik seçimi",font="helvetica 12",borderwidth=6)
grafik_label.place(x=25,y=575)

grafik_cesitleri=["Ayların yıllara göre değişimi(bar)","Yıl içinde mevsimlere göre değişim(bar)"]
grafik_kutu=Combobox(pencere,values=grafik_cesitleri,width=35)
grafik_kutu.place(x=25,y=600)

grafik_buton=Button(pencere,text="Seç",command=grafik_düzenle,font="helvetica 12",borderwidth=6)
grafik_buton.place(x=25,y=625)



interaktif_label1=Label(text="Kış Mevsimi Bar Grafiği",font="helvetica 12",borderwidth=6)
interaktif_label1.place(x=0,y=200)

interaktif_buton1=Button(pencere,text="Göster",background="blue",fg="white",command=interaktif1,font="helvetica 12",borderwidth=6)
interaktif_buton1.place(x=25,y=225)



interaktif_label2=Label(text="İlkbahar Mevsimi Bar Grafiği",font="helvetica 12",borderwidth=6)
interaktif_label2.place(x=0,y=275)

interaktif_buton2=Button(pencere,text="Göster",background="orange",fg="white",command=interaktif2,font="helvetica 12",borderwidth=6)
interaktif_buton2.place(x=25,y=300)



interaktif_label3=Label(text="Yaz Mevsimi Bar Grafiği",font="helvetica 12",borderwidth=6)
interaktif_label3.place(x=0,y=350)

interaktif_buton3=Button(pencere,text="Göster",background="black",fg="white",command=interaktif3,font="helvetica 12",borderwidth=6)
interaktif_buton3.place(x=25,y=375)



interaktif_label4=Label(text="Sonbahar Mevsimi Bar Grafiği",font="helvetica 12",borderwidth=6)
interaktif_label4.place(x=0,y=425)

interaktif_buton4=Button(pencere,text="Göster",background="red",fg="white",command=interaktif4,font="helvetica 12",borderwidth=6)
interaktif_buton4.place(x=25,y=450)



veri_label=Label(text="Sistemde kayıtlı dosya:Tüketici Fiyat Endeksi.xls",font="helvetica 12",borderwidth=6)
veri_label.place(x=0,y=50)

veri_text=Text(pencere,height=1,width=35)
veri_text.place(x=0,y=100)

veri_buton=Button(pencere,text="Onayla",command=text_bilgisi_al,font="helvetica 12",borderwidth=6)
veri_buton.place(x=0,y=125)

mainloop()
```
